```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
all_files=glob.glob(path+'/*xls')
```

```python
pd.read_excel(all_files[1],index_col=None,header=0)
```

```python
li=[]
for location in all_files:
    df=pd.read_excel(location,index_col=None,header=0)
    df=df[["Item",'2014-15']]
    df.loc[:,"States"]=location.split('\\')[-1].split('.')[0].split('_')[2]
    li.append(df)
li
```

```python
df=pd.concat(li,axis=0,ignore_index=True)
df=df.set_index("States")
```

```python
#cleaning data
df=df.dropna(axis=0,how='all')
df=df.dropna(axis=1,how='all')
df=df.drop("Puducherry",axis=0)
df=df.drop("Delhi",axis=0)
df=df.drop("Chandigarh",axis=0)
print("Drop Successful")
```

```python
#slicing data to select percapita GDP
per_capita=df.loc[(df.Item=='Per Capita GSDP (Rs.)')]
per_capita.sort_values("2014-15", axis = 0, ascending = True,inplace = True)
per_capita
```

```python
#Plot a Graph to Show PerCapita
%matplotlib inline
plt.style.use('ggplot')
per_capita.plot(kind='barh',figsize=(10,10),color=['Blue'],edgecolor='blue',alpha=0.8)
plt.ylabel('States', fontweight='bold', color = 'Orange', fontsize='18')
plt.xlabel('GDP per capita', fontweight='bold', color = 'Orange', fontsize='18')
plt.show()
plt.savefig('GDP_per_capita.png')
```

```python
#calculating Ratio
ratio=pd.DataFrame(per_capita)
ratio=ratio.reset_index()
d=ratio.iloc[0,[2]]
e=ratio.iloc[26,[2]]
f=e/d
f
```

```python
pst=df.loc[(df.Item == 'Primary') | (df.Item == 'Secondary')|(df.Item=='Tertiary')|(df.Item=='Per Capita GSDP (Rs.)')|(df.Item=="Population ('00)")]
GSDP=df.loc[(df.Item=='Gross State Domestic Product')]
pst_pivot = pst.pivot_table(values=['2014-15'],index='States',columns='Item', aggfunc='sum')
GSDP_pivot=GSDP.pivot_table(values=['2014-15'],index="States")
pst_percentages=pd.DataFrame(pst_pivot)
pst_percentages
```

```python
pst_percentages['%Primary']=pst_percentages.iloc[:,[2]].div(GSDP['2014-15'],axis=0).mul(100)
pst_percentages['%Secondary']=pst_percentages.iloc[:,[3]].div(GSDP['2014-15'],axis=0).mul(100)
pst_percentages['%Tertiary']=pst_percentages.iloc[:,[4]].div(GSDP['2014-15'],axis=0).mul(100)

pst_percentages
```

```python
correlation=pd.DataFrame(pst_percentages.iloc[:,[0,5]])
```

```python
#finding correlation
correlation=correlation.corr(method='pearson')
correlation
```

```python
#plottig correlation
sns.pairplot(correlation)
plt.savefig('coorelation.png')
```

```python
#plotting percentage contribution of sectors
pst_bar=pst_percentages.iloc[:,[5,6,7]]
colors=['Blue','green','orange']
barWidth =0.5
pst_bar.plot(kind='bar',stacked="True",figsize=(20,10),color=colors,width=barWidth, edgecolor='white')
plt.legend( loc='upper left')

plt.ylabel(" %Contribution of Each Sector to GSDP", fontsize=14, labelpad=15)
plt.xlabel("Month Year", fontsize=16, labelpad=15)
plt.title("Percentage Contribution of the Primary, Secondary and Tertiary sectors as a percentage of the total GDP", fontsize=18, y=1.02)
plt.savefig('contributionof sectors.png')
```
